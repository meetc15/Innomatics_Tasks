Unique Words in a Sentence

```python
# Take input from user
sentence = input("Enter sentence: ")

# Convert sentence to lowercase so 'Python' and 'python' are treated the same
sentence = sentence.lower()

# Split the sentence into words using spaces
words = sentence.split()

# Convert list of words into a set to remove duplicates
unique_words = set(words)

# Count number of unique words
count_unique = len(unique_words)

# Display results
print("Unique words count:", count_unique)
print("Unique words:", unique_words)
```

Highest Salary from Employee Data

```python
# Employee salary data stored in a dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find the employee with the highest salary
# max() compares dictionary keys based on their values using employees.get
highest_employee = max(employees, key=employees.get)

# Print the employee name and their salary
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

Find Maximum and Minimum Values

```python
# List of numbers
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with the first element
max_value = numbers[0]
min_value = numbers[0]

# Loop through the list to compare values
for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

# Display results
print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

Count Products Above a Price Threshold

```python
# List of product prices
prices = [450, 1200, 899, 1500, 300]

# Counter for products above 1000
count = 0

# Loop through prices and check condition
for price in prices:
    if price > 1000:
        count += 1

# Display result
print("Products above 1000:", count)
```

Calculate Attendance Percentage

```python
# Attendance list
attendance = ["P", "P", "A", "P", "P"]

# Count present days
present_count = attendance.count("P")

# Calculate percentage
percentage = (present_count / len(attendance)) * 100

# Display result
print("Attendance Percentage:", percentage)
```

Count Character Frequency

```python
# Input string
text = "pythonp"

# Dictionary to store character frequency
frequency = {}

# Count each character
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

# Display result
print(frequency)
```

Check if a Key Exists in a Dictionary

```python
# Dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Key to check
key = "Ravi"

# Check if key exists
if key in employees:
    print("Employee exists")
else:
    print("Employee not found")
```

Calculate Average Marks

```python
# List of marks
marks = [70, 75, 80, 65]

# Calculate average
average = sum(marks) / len(marks)

# Display result
print("Average Marks:", average)
```
