Mobile Recharge Validation System

```python
# Valid recharge plans
VALID_PLANS = [199, 299, 399, 599]

# Function to validate recharge
def validate_recharge(amount):
    if amount < 50:
        print("Minimum recharge amount is ₹50")
        return False

    if amount not in VALID_PLANS:
        print("Invalid plan selected")
        print("Available plans:", VALID_PLANS)
        return False

    return True


# Main program with retry system
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))

        if validate_recharge(amount):
            print("Recharge Successful!")
            break
        else:
            print("Please try again...\n")

    except ValueError:
        print("Please enter numbers only\n")
```

Inventory Reorder Alert System

```python
# inventory dictionary
inventory = {
    "Rice": 40,
    "Wheat": 12,
    "Sugar": 25,
    "Salt": 8,
    "Oil": 15,
    "Milk Powder": 5
}

# function to check stock
def check_stock(product, quantity):
    if quantity < 15:
        print("Reorder Alert -", product, "Stock:", quantity)
    else:
        print("Stock OK -", product, "Stock:", quantity)

# loop through dictionary
for product, quantity in inventory.items():
    check_stock(product, quantity)
```

Student Result Processing System

```python
# function to process result
def process_result(marks):
    total = 0
    
    # calculate total using loop
    for m in marks:
        total += m
    
    average = total / len(marks)

    if average >= 50:
        return "Pass", average
    else:
        return "Fail", average


# example data
student_marks = [60, 45, 70, 55, 40]

# call function
result, avg = process_result(student_marks)

print("Average:", avg)
print("Result:", result)
```

Cab Fare Estimator with Retry Option

```python
# function to calculate fare
def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)

    if peak.lower() == "yes":
        fare = fare + (0.25 * fare)

    return fare


# retry system
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")

        if distance <= 0:
            print("Invalid distance. Try again.\n")
            continue

        total_fare = calculate_fare(distance, peak)
        print("Total Fare:", round(total_fare, 2))
        break

    except ValueError:
        print("Please enter numeric value for distance.\n")
```

Employee Attendance Eligibility Checker

```python
# function to check eligibility
def check_eligibility(attendance):
    present_days = 0

    # count present days
    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible", percentage
    else:
        return "Not Eligible", percentage


# example attendance record
record = ["P", "P", "A", "P", "P", "A", "P", "P"]

status, percent = check_eligibility(record)

print("Attendance Percentage:", percent)
print("Status:", status)
```

Password Strength Checker

```python
# function to check password strength
def check_password(password):
    has_digit = False
    has_special = False

    # check characters
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# input from user
pwd = input("Enter password: ")

result = check_password(pwd)
print(result)
```
